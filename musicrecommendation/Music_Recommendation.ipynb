# Music Recommendation
Music recommender system can suggest songs to users based on their listening pattern.

### import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import Recommenders as Recommenders
```

```python

    song_df_1=pd.read_csv('triplets_file.csv')
    song_df_1.head()
```

```python
len(song_df_1)
```

```python
song_df_2=pd.read_csv('song_data.csv')
song_df_2.head()
```

```python
len(song_df_2)
```

### Merging two dataframes and dropping duplicates(if any)

```python
song_df=pd.merge(song_df_1,song_df_2.drop_duplicates(['song_id']),on='song_id',how='left')
song_df.head()
```

```python
len(song_df)
```

### Data Preprocessing

```python
song_df['song']=song_df['title']+"-"+song_df['artist_name']
song_df.head()
```

```python
song_df.tail()
```

### Data Visualisation

#### Correlation between features

```python
song_df.columns
```

```python
song_corr=song_df.corr()
print(song_corr)
```

#### Most listened songs

```python
#Top 15
plt.style.use('seaborn')
sns.barplot(x=song_df['song'].value_counts()[:15].values,y=song_df['song'].value_counts()[:15].index)
plt.xlabel('No. of persons')
plt.ylabel('Song title')
plt.title('Most listened songs')
```

#### Reducing the size of dataset

```python
song_df=song_df.head(100000)
```

#### Sum of listen_count of different songs

```python
song_grouped=song_df.groupby(['song']).agg({'listen_count':'count'}).reset_index()
song_grouped.head()
```

#### Most listened songs and their corresponding percentage

```python
grouped_sum=song_grouped['listen_count'].sum()
song_grouped['percentage']=(song_grouped['listen_count']/grouped_sum)*100
song_grouped.sort_values(['listen_count','song'],ascending=[0,1])
```

#### Recommendation Engine based on Popularity of Songs

```python
pr=Recommenders.popularity_recommender_py()
```

```python
pr.create(song_df,'user_id','song')
```

```python
pr.recommend(song_df['user_id'][5])
```

```python
pr.recommend(song_df['user_id'][10])
```
